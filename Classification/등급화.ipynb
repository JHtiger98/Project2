```python
pop_density = pd.read_csv('인구밀도.csv')

pop_density_2014 = pop_density.iloc[2:,[1,5]]

pop_density_2014['2014.2'] = pop_density_2014['2014.2'].replace('-',0)
pop_density_2014['2014.2'] = pop_density_2014['2014.2'].astype('int64')

pop_density_2014 = pop_density_2014.groupby(['동별(2)']).agg(sum = ('2014.2', 'sum'))

pop_density_2014 = pop_density_2014.reset_index()

pop_density_2014.columns = ('자치구', '2014')

pop_density_2014 = pop_density_2014.set_index('자치구')

pop_density_2014
```

```python
pop_density = pd.read_csv('인구밀도.csv')

pop_density_2015 = pop_density.iloc[2:,[1,8]]

pop_density_2015['2015.2'] = pop_density_2015['2015.2'].replace('-',0)
pop_density_2015['2015.2'] = pop_density_2015['2015.2'].astype('int64')

pop_density_2015 = pop_density_2015.groupby(['동별(2)']).agg(sum = ('2015.2', 'sum'))

pop_density_2015 = pop_density_2015.reset_index()

pop_density_2015.columns = ('자치구', '2015')

pop_density_2015 = pop_density_2015.set_index('자치구')

pop_density_2015
```

```python
marriage['2014'] = marriage['2014'].astype('int64')
marriage['2015'] = marriage['2015'].astype('int64')
```

```python
marriage['marriage'] = (marriage['2014'] + marriage['2015']) / 2
marriage = marriage.iloc[:,[0,3]]
marriage.columns = ('자치구', 'marriage')
marriage = marriage.set_index('자치구')
marriage
```

```python
labels = ['5','4','3','2','1']
marriage_bins = pd.cut(marriage['marriage'], 5, labels=labels)
marriage_rank = pd.DataFrame(marriage_bins)
marriage_rank
```

```python
marriage_rank_increase = pd.merge(increase_store, marriage_rank, on='자치구', how='inner').reset_index().iloc[:,1:].set_index('점포코드')
marriage_rank_increase
```

```python
marriage_rank_decrease = pd.merge(decrease_store, marriage_rank, on='자치구', how='inner').reset_index().iloc[:,1:].set_index('점포코드')
marriage_rank_decrease
```
